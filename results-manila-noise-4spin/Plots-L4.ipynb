```python
import matplotlib.pyplot as plt
import numpy as np

import sys
sys.path.insert(0, '../src')
from plots import *

L = 4

true_energies = np.load(f"../exact_gse_L{L}.npy")
true_d2e = np.load(f"../exact_d2e_L{L}.npy")
true_fid = np.load(f"../exact_fs_L{L}.npy")
```

```python
en_unmit = np.load(f"results-{L}spin_unmitigated_ground_state_energy.npy")

en_unitary13 = np.load(f"results-{L}spin_unitary13_ground_state_energy.npy")
en_unitary135 = np.load(f"results-{L}spin_unitary135_ground_state_energy.npy")
en_unitary1357 = np.load(f"results-{L}spin_unitary1357_ground_state_energy.npy")

en_cnot13 = np.load(f"results-{L}spin_cnot13_ground_state_energy.npy")
en_cnot135 = np.load(f"results-{L}spin_cnot135_ground_state_energy.npy")
en_cnot1357 = np.load(f"results-{L}spin_cnot1357_ground_state_energy.npy")
```

```python
d2e_unmit = np.load(f"results-{L}spin_unmitigated_second_energy_derivative.npy")

d2e_unitary13 = np.load(f"results-{L}spin_unitary13_second_energy_derivative.npy")
d2e_unitary135 = np.load(f"results-{L}spin_unitary135_second_energy_derivative.npy")
d2e_unitary1357 = np.load(f"results-{L}spin_unitary1357_second_energy_derivative.npy")

d2e_cnot13 = np.load(f"results-{L}spin_cnot13_second_energy_derivative.npy")
d2e_cnot135 = np.load(f"results-{L}spin_cnot135_second_energy_derivative.npy")
d2e_cnot1357 = np.load(f"results-{L}spin_cnot1357_second_energy_derivative.npy")
```

```python
fs_unmit = np.load(f"results-{L}spin_unmitigated_fidelity_susceptibility.npy")

fs_unitary13 = np.load(f"results-{L}spin_unitary13_fidelity_susceptibility.npy")
fs_unitary135 = np.load(f"results-{L}spin_unitary135_fidelity_susceptibility.npy")
fs_unitary1357 = np.load(f"results-{L}spin_unitary1357_fidelity_susceptibility.npy")

fs_cnot13 = np.load(f"results-{L}spin_cnot13_fidelity_susceptibility.npy")
fs_cnot135 = np.load(f"results-{L}spin_cnot135_fidelity_susceptibility.npy")
fs_cnot1357 = np.load(f"results-{L}spin_cnot1357_fidelity_susceptibility.npy")
```

```python
en_unitary = [en_unmit, en_unitary13, en_unitary135, en_unitary1357]
en_cnot = [en_unmit, en_cnot13, en_cnot135, en_cnot1357]

d2e_unitary = [d2e_unmit, d2e_unitary13, d2e_unitary135, d2e_unitary1357]
d2e_cnot = [d2e_unmit, d2e_cnot13, d2e_cnot135, d2e_cnot1357]

fs_unitary = [fs_unmit, fs_unitary13, fs_unitary135, fs_unitary1357]
fs_cnot = [fs_unmit, fs_cnot13, fs_cnot135, fs_cnot1357]
```

## Ground state energies

```python
fig, ax = plt.subplots(1, 2, figsize=(20,6), sharey=True)
make_scatter_plot(ax, true_energies, np.array(en_unitary), np.array(en_cnot), "Ground state energy")    
plt.suptitle(f"$L={L}$, ground state energy", fontsize=20)
plt.tight_layout()
plt.savefig(f"l{L}_gse.pdf")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20,6), sharey=True)
make_mean_absolute_error_plot(ax[0], true_energies, en_unitary, en_cnot)
make_absolute_error_mean_plot(ax[1], true_energies, en_unitary, en_cnot, legend=False)
plt.suptitle(f"$L={L}$,\t ground state energy", fontsize=20)
plt.tight_layout()
plt.savefig(f"l{L}_gse_error.pdf")
```

## Second energy derivatives

```python
fig, ax = plt.subplots(1, 2, figsize=(20,6), sharey=True)
make_scatter_plot(ax, true_d2e, d2e_unitary, d2e_cnot, "-$(\partial^{2}$E(r)/$\partial$r$^{2})$/L")
plt.suptitle(f"L={L}, second energy derivative", fontsize=20)
plt.tight_layout()
plt.savefig(f"l{L}_d2e.pdf")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20,6), sharey=True)
make_mean_absolute_error_plot(ax[0], true_d2e, d2e_unitary, d2e_cnot)
make_absolute_error_mean_plot(ax[1], true_d2e, d2e_unitary, d2e_cnot, legend=False)
plt.suptitle(f"$L={L}$,\t second energy derivative", fontsize=20)
plt.tight_layout()
plt.savefig(f"l{L}_d2e_error.pdf")
```

## Fidelity susceptibilities

```python
fig, ax = plt.subplots(1, 2, figsize=(20,6), sharey=True)
make_scatter_plot(ax, true_fid, fs_unitary, fs_cnot, "Fidelity susceptibility / L")    
plt.suptitle(f"L={L}, fidelity susceptibility", fontsize=20)
plt.tight_layout()
plt.savefig(f"l{L}_fs.pdf")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20,6), sharey=True)
make_mean_absolute_error_plot(ax[0], true_fid, fs_unitary, fs_cnot)
make_absolute_error_mean_plot(ax[1], true_fid, fs_unitary, fs_cnot, legend=False)
plt.suptitle(f"$L={L}$,\tFidelity susceptibility/$L$", fontsize=20)
plt.tight_layout()
plt.savefig(f"l{4}_fs_error.pdf")
```
